```python
import pandas as pd
import numpy as np
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
# Load the dataset
data = pd.read_csv("AirPassengers.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
# Convert the data into datatime datatype. If we don't set the index and change the datatype for the data, it will throw error.
data['Month'] = pd.to_datetime(data['Month'])
data.set_index('Month', inplace=True)
```

```python
result = seasonal_decompose(data, model='multiplicative')
```

```python
result.plot();
```

```python
result_1 = seasonal_decompose(data, model='additive')
```

```python
result_1.plot();
```

However, we will choose the multiplicative decomposition as there is an increase in the trend and also an increase in the seasonality of the data
