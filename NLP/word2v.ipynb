```python
import gensim
from gensim.models import Word2Vec, KeyedVectors
```

```python
model = KeyedVectors.load_word2vec_format('GoogleNews-vectors-negative300.bin.gz',binary=True,limit=500000)
```

```python
model['man'] # This is numerical vector represents word man. Output is 300 numbers
```

```python
model['glass'] # this is the numerical vector that represents word glass
```

```python
model['man'].shape
```

```python
model.most_similar('man') #word in similar to man 
```

```python
model.most_similar('facebook')
```

```python
model.similarity('man','woman') #How similar 2 words are
```

```python
model.doesnt_match(['Java','Python','Elephant'])
```

```python
vec = model['king'] - model['man'] + model['woman']
model.most_similar([vec])
```

Intuition: Use Word2vec algorithm to find the sematics between words.

```python
# word used: king queen woman girl boy man water
model['king'][:50]
```

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
words = ['queen','woman','girl','boy','man','king','water']
embeddings = np.array([model['queen'][:50], 
                       model['woman'][:50], 
                       model['girl'][:50], 
                       model['boy'][:50], 
                       model['man'][:50],
                       model['king'][:50],  
                       model['water'][:50]])
df = pd.DataFrame(embeddings, index=words)
```

```python
# create a heatmap
plt.figure(figsize =(12,12))
heatmap = sns.heatmap(df,cmap='crest')
plt.show();
```
