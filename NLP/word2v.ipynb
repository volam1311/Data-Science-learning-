```python
df['label_num'] = df['label'].map(
    {
        "Fake":0,
        "Real":1
    }
)
```

```python
df.head()
```

Before predicting the text, we will do some preprocessing

```python
import spacy
nlp = spacy.load('en_core_web_lg')
```

```python
def preprocess_and_vectorize(text):
    doc = nlp(text)
    filtered_token = []
    for token in doc:
        if token.is_punct or token.is_stop:
            continue
        filtered_token.append(token.lemma_)

    return wv.get_mean_vector(filtered_token)
```

```python
preprocess_and_vectorize('What the hell is that')
```

The easiest method for a sentence embedding is to take the average of the vectors in the sentence

```python
v1 = wv.get_mean_vector(['worry','understand'])
```

```python
v1.shape
```

When you use the get_mean_vector method, it will always give you the normalize mean vector

```python
df['vector'] = df['Text'].apply(lambda text: preprocess_and_vectorize(text))
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test, y_train, y_test = train_test_split(df.vector.values, df.label_num,test_size=0.2,random_state=42,stratify=df.label_num)
```

```python
x_train_2d = np.stack(x_train)
x_test_2d = np.stack(x_test)
```

Change the array inside array to only array

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report

clf = GradientBoostingClassifier()
clf.fit(x_train_2d,y_train)
```

```python
y_pred = clf.predict(x_test_2d)
print(classification_report(y_test,y_pred))
```
